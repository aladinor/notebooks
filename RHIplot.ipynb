# RHI

```python
path = "/home/alfonso/Downloads"
filename = f"{path}/cfrad.20211011_201733.023_to_20211011_201745.299_DOW8_RHI.nc"
filename = f"{path}/cfrad.20211011_223602.712_to_20211011_223612.091_DOW8_RHI.nc"
```
